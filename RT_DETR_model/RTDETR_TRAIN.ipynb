```python
from ultralytics import RTDETR
import os
import glob
import numpy as np
from collections import defaultdict

def calc_pa_stats(model, test_images_dir, conf_threshold=0.25):
    """
    Calculate per-image presence/absence metrics for RT-DETR,
    reporting both Micro and Macro averages.
    """
    test_labels_dir = test_images_dir.replace("/images", "/labels")

    if not os.path.exists(test_labels_dir):
        print(f"Warning: Labels directory not found at {test_labels_dir}")
        print("Ensure your folder structure is '.../images' and '.../labels'")
        return

    class_names = model.names
    num_classes = len(class_names)

    # Initialize metrics dictionaries
    TP = defaultdict(int)
    FP = defaultdict(int)
    FN = defaultdict(int)

    image_extensions = ["*.jpg", "*.jpeg", "*.png"]
    image_files = []
    for ext in image_extensions:
        image_files.extend(glob.glob(os.path.join(test_images_dir, ext)))

    print(f"Processing {len(image_files)} images for Presence/Absence metrics...")

    for img_path in image_files:
        base = os.path.splitext(os.path.basename(img_path))[0]
        label_path = os.path.join(test_labels_dir, base + ".txt")

        # Get ground truth classes for current image
        gt_classes = set()
        if os.path.exists(label_path):
            with open(label_path, "r") as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) > 0:
                        gt_classes.add(int(float(parts[0])))

        # Run RT-DETR inference
        results = model.predict(img_path, stream=False, verbose=False)
        preds = results[0].boxes

        # Filter predictions by confidence threshold
        pred_classes = set(
            int(c) for c, conf in zip(preds.cls.cpu().numpy(), preds.conf.cpu().numpy())
            if conf > conf_threshold
        )

        # Compute TP, FP, FN per class
        for c in range(num_classes):
            is_present_gt = c in gt_classes
            is_present_pred = c in pred_classes

            if is_present_gt and is_present_pred:
                TP[c] += 1
            elif not is_present_gt and is_present_pred:
                FP[c] += 1
            elif is_present_gt and not is_present_pred:
                FN[c] += 1

    # --- Print Report ---
    print("\n" + "="*80)
    print(f"Per-Class Image-Level Metrics (Conf > {conf_threshold})")
    print("="*80)

    # Variables for Micro Calculation (Global sums)
    overall_tp = overall_fp = overall_fn = 0

    # Lists for Macro Calculation (Per-class scores)
    macro_p_list = []
    macro_r_list = []
    macro_f1_list = []

    # Print Header
    print(f"{'Class Name':<20} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'(TP/FP/FN)':<15}")
    print("-" * 80)

    for c, name in class_names.items():
        tp, fp, fn = TP[c], FP[c], FN[c]

        # Calculate Per-Class metrics
        prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
        rec = tp / (tp + fn) if (tp + fn) > 0 else 0.0
        f1 = (2 * prec * rec) / (prec + rec) if (prec + rec) > 0 else 0.0

        print(f"{name:<20} {prec:<10.3f} {rec:<10.3f} {f1:<10.3f} ({tp}/{fp}/{fn})")

        # Accumulate for Micro
        overall_tp += tp
        overall_fp += fp
        overall_fn += fn

        # Store for Macro
        macro_p_list.append(prec)
        macro_r_list.append(rec)
        macro_f1_list.append(f1)

    print("-" * 80)

    # --- Calculate Averages ---

    macro_prec = np.mean(macro_p_list)
    macro_rec = np.mean(macro_r_list)
    macro_f1 = np.mean(macro_f1_list)

    micro_prec = overall_tp / (overall_tp + overall_fp) if (overall_tp + overall_fp) > 0 else 0
    micro_rec = overall_tp / (overall_tp + overall_fn) if (overall_tp + overall_fn) > 0 else 0
    micro_f1 = (2 * micro_prec * micro_rec) / (micro_prec + micro_rec) if (micro_prec + micro_rec) > 0 else 0

    print(f"{'MACRO AVERAGE':<20} {macro_prec:<10.3f} {macro_rec:<10.3f} {macro_f1:<10.3f}")
    print(f"{'MICRO AVERAGE':<20} {micro_prec:<10.3f} {micro_rec:<10.3f} {micro_f1:<10.3f}")
    print("="*80 + "\n")


# --- Main Execution Block ---

WEIGHTS_PATH = "/content/drive/MyDrive/Colab Notebooks/rtdetr_output/rtdetr_run_01/weights/best.pt"
TEST_IMAGES_DIR = "/content/drive/MyDrive/Colab Notebooks/test/images"

if __name__ == "__main__":
    if not os.path.exists(WEIGHTS_PATH):
        print(f"ERROR: Weights not found at: {WEIGHTS_PATH}")
    elif not os.path.exists(TEST_IMAGES_DIR):
        print(f"ERROR: Test images not found at: {TEST_IMAGES_DIR}")
    else:
        print(f"Loading RT-DETR model from: {WEIGHTS_PATH}")
        model = RTDETR(WEIGHTS_PATH)

        calc_pa_stats(model, TEST_IMAGES_DIR, conf_threshold=0.4)
```
